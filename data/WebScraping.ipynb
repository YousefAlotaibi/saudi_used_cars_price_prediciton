```python
import cloudscraper
from bs4 import BeautifulSoup 
import requests
import pandas as pd
from random import randint
from time import sleep
import numpy as np
```

```python
headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:103.0) Gecko/20100101 Firefox/103.0'
}
```

```python
links = []
car_name = []
model_year = []
km = []
cars = []
transmission = []
fuel = []
color = []
No_doors = []
Engine_capacity = []
Drive_type = []
No_cylinders = []
Accident_history = []
price = []
```

##### Each time I try to scrape the data an error rises and I would have only 5800-5900 records out of 9000 scraped, So I changed the range for pages 3 times first from 1-500 pages, second from 250-500 pages and then from 400-500 pages.

```python
scraper = cloudscraper.create_scraper()

for x in range(1, 500):
    r = scraper.get(f'https://ksa.yallamotor.com/used-cars/search?page={x}&sort=updated_desc')
    soup = BeautifulSoup(r.text, 'html.parser')
    cars = soup.select('.singleSearchCard')
    sleep(randint(2,10))
    for car in cars:
        url1 = car.select_one('a.black-link')['href']     

        session_object = requests.Session()
        result = session_object.get(f'https://ksa.yallamotor.com/{url1}', headers=headers)
        soup = BeautifulSoup(result.text, 'lxml')

        name = soup.find('h1', class_="font24")
        car_name.append(name.text)

        p = soup.find('span', class_="font28 font-b")
        price.append(p.text)

        y = soup.find_all('div', class_="font14 text-center font-b m2t")[0]
        model_year.append(y.text)
        
        k = soup.find_all('div', class_="font14 text-center font-b m2t")[1]
        km.append(k.text)

        t = soup.find_all('div', class_="font14 text-center font-b m2t")[3]
        transmission.append(t.text)
        
        f = soup.find_all('div', class_="font14 text-center font-b m2t")[4]
        fuel.append(f.text)

        c = soup.find_all('div', class_="font14 text-center font-b m2t")[5]
        color.append(c.text)

        d = soup.find_all('div', class_="font14 text-center font-b m2t")[7]
        No_doors.append(d.text)

        e = soup.find_all('div', class_="col is-5 p0 font-b")[3]
        Engine_capacity.append(e.text)

        dt = soup.find_all('div', class_="col is-5 p0 font-b")[6]
        Drive_type.append(dt.text)

        nc = soup.find_all('div', class_="col is-5 p0 font-b")[10]
        No_cylinders.append(nc.text)

        ah = soup.find_all('div', class_="col is-5 p0 font-b")[12]
        Accident_history.append(ah.text)
```

```python
cars_data = pd.DataFrame()
```

```python
cars_data['car_name'] = car_name
cars_data['model_year'] = model_year
cars_data['km'] = km
cars_data['transmission'] = transmission
cars_data['fuel'] = fuel
cars_data['color'] = color
cars_data['no_doors'] = No_doors
cars_data['engine_capacity'] = Engine_capacity
cars_data['drive_type'] = Drive_type
cars_data['data'] = No_cylinders
cars_data['Accident_history'] = Accident_history
cars_data['price'] = price
```
